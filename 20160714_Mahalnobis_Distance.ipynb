# マハラビノス距離

* マハラビノス距離では、ある敷居値を自分自身で設定し、敷居値を超えたレコードを異常値として判定することができる
* 今回は敷居値を設けずにグラフとして図示する。

```python
# -*- coding: utf-8 -*-
import numpy as np
import scipy as sc
from scipy import linalg
from scipy import spatial 
import scipy.spatial.distance 
import pandas as pd
import matplotlib.pyplot as plt 
#import matplotlib.font_manager import pylab

ROW = 6
COLUMN = 10
xx=["a","b","c","d","e","f"]
csv = pd.read_csv('./datas/test.txt',header=0)
csv.index=xx
csv.head(6)
```

今後扱う変数も定義する

```python
# row:行,column:列,ave:平均,vcm:分散共分散行列 row = []
column = []
row=[]
ave = [0.0 for i in range(ROW)]
vcm = np.zeros((COLUMN, ROW, ROW)) 
diff = np.zeros((1, ROW))
mahal = np.zeros(COLUMN)
tmp = np.zeros(ROW)
```

#  欠損値の削除
* 一般的にデータ郡は欠損値がある。

```python
trans_data = csv.dropna(axis=1) 
print(trans_data)
```

# 多次元配列でROW, COLUMを格納する

```python
row=[]
column=[]
for i in range(ROW):
    #print (list(trans_data.ix[i]))
    row.append(list(trans_data.ix[i])) 
print(row)

#列 連結
for i in range(0, COLUMN):
    column.append(list(trans_data.ix[:, i])) 
print(column)
```

# 平均値計算

```python
for i in range(ROW):
    # スライスという技法
    ave[i] = np.average(row[i][1:len(row[i])]) 
print(ave)
```

# 分散共分散行列の計算

```python
# Numpyのメソッドを使う。
column = np.array([column])
ave = np.array(ave)
# 分散共分散行列を求める。np.swapaxes()で軸を変換することで求めることができる
for i in range(COLUMN):
    diff = (column[0][i] - ave)
    diff = np.array([diff])
    vcm[i] = (diff * np.swapaxes(diff, 0, 1)) / COLUMN
print(vcm)
```
