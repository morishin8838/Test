# seaborn

seabornはmatplotlibをベースにしている。見目麗しいみたい。

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
# デフォルトでバンドルされている、それを使って分析する
x = np.random.normal(size=100)      #ランダムデータ作成
titanic = sns.load_dataset("titanic") ##kaggleで有名な、タイタニック号の生死者データ
tips = sns.load_dataset("tips")         ## お店の食事時間と会計総額とチップの関係のデータ
iris = sns.load_dataset("iris")            ## Rでお馴染みのアヤメの統計データ
```

```python
# ヒストグラム
> kdeは密度近似関数の描画のオプション、binsはx軸の刻み目の指定です。
rugに関してはTrueにすると何が起こるか試してみてください。
```

```python
sns.distplot(x, kde=True, rug=False, bins=10)
```

# 散布図

## jointplot
x軸とy軸の数値の分布を美しく表示する

```python
sns.jointplot('sepal_width', 'petal_length', data=iris)
```

## pairplot

データに含まれる全ての数値変数同士について簡単に見ることが出来る

```python
sns.pairplot(iris)
```
