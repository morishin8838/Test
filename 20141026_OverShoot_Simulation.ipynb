```python
from pylab import *
from control.matlab import *
import numpy as np
from scipy.signal import lti
from scipy.signal import step as p_step

tt=arange(0, 1.2, 0.02)
tt_bslm=arange(0, 0.05, 0.001)
```

```python
#1) Step応答解析 OS:行き過ぎ量(ピーク値と定常値との割合) Tr:立ち上げり時間 90%に至る時間 Ts:整定時間 2%に入る時間
def step_info(t,yout):
    print "OS: %f%s"%((yout.max()/yout[-1]-1)*100,'%')
    print "Tr: %fs"%(t[next(i for i in range(0,len(yout)-1) if yout[i]>yout[-1]*.90)]-t[0])
    print "Ts: %fs"%(t[next(len(yout)-i for i in range(2,len(yout)-1) if abs(yout[-i]/yout[-1])>1.02)]-t[0])
```

## システム定義

```python
#sys = s / (s^2 + 4*s + 20)
m=0.01;     # Gain Mass gain block
c=10;       # Damping gain block　　　粘性
k=50000;    # Stiffness gain block ばね定数
num = [1,1]; 
den = [m, c, k]; 
sys = tf(num,den)
sys_c=feedback(sys,1)
sys = sys_c * 10
#sys2 = negate(sys1) # Same as sys2 = -sys1.
#sys3 = parallel(sys1, sys2) # Same as sys3 = sys1 + sys2.
#sys3 = series(sys1, sys2) # Same as sys3 = sys2 * sys1.
```
